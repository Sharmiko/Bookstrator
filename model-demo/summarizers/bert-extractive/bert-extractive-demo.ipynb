```
# store csv in dictionary
fables = {}

with open('data/aesops_out.csv', 'r') as infile:
  csv_reader = csv.reader(infile)
  for line in csv_reader:
    fables[line[0]] = line[1]
```

```
from summarizer import Summarizer
```

```
summarizer = Summarizer()
```

```
def summarize(text: str, min_length=60, max_length=600):
  return ''.join(summarizer(text, max_length=max_length))
```

```
def summarize_fable(fable_name: str, min_length=60, max_length=600):
  fable_body = fables[fable_name]
  return fable_body, summarize(fable_body, min_length=min_length, max_length=max_length)
```

```
body, summary = summarize_fable('The Wolf and the Lamb', max_length=600)
```
